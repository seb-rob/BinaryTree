#### Take input from the user and create Binary Tree and print nodes maintaining relations

```python
class BinaryTree_3:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
# write function to print complete binary Tree
def printBT_3(root):
    if root is None:
        return
    print(root.data, end=":")
    if root.left is not None:
        print("L", root.left.data, end=" ")
    if root.right is not None:
        print("R", root.right.data, end="")
    print()
    printBT_3(root.left)
    printBT_3(root.right)
    
# write funtion through which user can give input for binary tree
def inputBT():
    treeData = int(input())
    # if user enter -1 that means user doesn't want to have root nodes child
    if treeData == -1:
        return None
    # create node
    cur_root = BinaryTree_3(treeData)
    # again take input for root's left
    leftVal = inputBT()
    # agian take input for root's right
    rightVal = inputBT()
    # establish connection
    cur_root.left = leftVal
    cur_root.right = rightVal
    return cur_root

root = inputBT()
printBT_3(root)
```

#### create binary tree, count number of nodes present in the tree

```python
class BinaryTree_5:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# take input for binary tree
def inputBT_5():
    data = int(input())
    if data == -1:
        return
    new_node = BinaryTree_5(data)
    left_tree = inputBT_5()
    right_tree = inputBT_5()
    
    new_node.left = left_tree
    new_node.right = right_tree
    return new_node

# print Binary tree
def printBT_5(root):
    if root is None:
        return
    print(root.data, end=":")
    if root.left is not None:
        print("L", root.left.data, end=",")
    if root.right is not None:
        print("R", root.right.data, end="")
    print()
    printBT_5(root.left)
    printBT_5(root.right)
    
# count number of nodes present in the tree
def countNumberofNodes(root):
    if root is None:
        return 0
    return 1 + countNumberofNodes(root.left) + countNumberofNodes(root.right)

root = inputBT_5()
printBT_5(root)
numberofNodes = countNumberofNodes(root)
print("number of nodes present in the tree is: ", numberofNodes)
```
